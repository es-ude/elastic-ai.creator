```python
import torch
from torch import nn
import matplotlib.pyplot as plt
from elasticai.creator.nn.fixed_point import Tanh
```

```python
x = torch.linspace(-4, 4, 1000)
tanh1 = nn.Tanh()
tanh2 = Tanh(total_bits=8, frac_bits=5, num_steps=2**8, sampling_intervall=(-4, 3.96875))

y1 = tanh1(x)
y2 = tanh2(x)
```

```python
# Plot the functions
plt.figure(figsize=(8, 6))
plt.plot(x.numpy(), y1.numpy(), label='Tanh', linewidth=2)
plt.plot(x.numpy(), y2.numpy(), label='Precomputed Tanh', linewidth=2)

# Labels and title
plt.xlabel(r'x')
plt.ylabel(r'tanh(x)')
plt.title(r'Comparison of full precision and $Q_{8.5}$ Tanh')
plt.legend()
plt.grid(True)

# Show plot
plt.savefig(f'precomputed_tanh.eps', format='eps')
plt.show()
```
